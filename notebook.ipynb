```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('dataset/Top 1000 Sekolah Clean.xlsx', decimal=',')
df.head()
```

```python
lokasi = df['Kota/Kab - Prov'].str.split(" - ", n = 1, expand = True)
```

```python
df['Kota/Kab'] = lokasi[0]
df['Provinsi'] = lokasi[1]
df.head()
```

```python
del(df['Kota/Kab - Prov'])
df.head()
```

```python
cols = list(df.columns)

df = df[cols[0:2] + [cols[-2]] + [cols[-1]] + cols[2:15]]
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
import re
sman = df[df['Nama Sekolah'].str.contains('SMAN', flags=re.IGNORECASE)]
sman.info()
```

```python
sman_n = df[df['Nama Sekolah'].str.contains('SMA N', flags=re.IGNORECASE)]
sman_n.info()
```

```python
df = pd.concat([sman, sman_n], ignore_index=True)
df.info()
```

```python
df.sort_values(by='Nilai Total', ascending=False).head(10)
```

```python
top_sman1 = df.sort_values(by='Nilai Total', ascending=False).head(5)
top_sman1.head()
```

```python
plt.figure(figsize=(8,5))
bar_chart = plt.bar(top_sman1['Nama Sekolah'], top_sman1['Nilai Total'], color='Green', width=0.7)

plt.title('Top 5 SMAN', size=16, pad=15)
plt.ylabel('Nilai Total', size=14)
plt.xticks(size=12, rotation=325, ha='left')
plt.yticks(size=12)
plt.ylim(ymin=600 ,ymax=650)
plt.bar_label(bar_chart, padding=10, size=14)
plt.show()
```

```python
per_prov = df['Provinsi'].value_counts().rename_axis('Provinsi').reset_index(name='Jumlah SMAN')
per_prov
```

```python
top_prov = df['Provinsi'].value_counts().nlargest(5).rename_axis('Provinsi').reset_index(name='Jumlah SMAN')
top_prov
```

```python
plt.figure(figsize=(8,6))
bar_chart = plt.bar(top_prov['Provinsi'], top_prov['Jumlah SMAN'], color='red', width=0.7)

plt.title('Top 5 Provinsi', size=16, pad=15)
plt.ylabel('Jumlah Sekolah', size=14)
plt.xticks(size=12, rotation=325, ha='left')
plt.yticks(size=12)
plt.ylim(ymax=180)
plt.bar_label(bar_chart, padding=10, size=14)
plt.show()
```

```python
df_prov = df.groupby('Provinsi').agg({'Nilai Total': 'mean', 'Tes Potensi Skolastik':'mean', 'TKA Saintek':'mean', 'TKA Soshum':'mean'}).reset_index()
df_prov
```

```python
df_prov.sort_values(by=['Nilai Total'], ascending=False)
```

```python
df_prov[['Tes Potensi Skolastik', 'TKA Saintek', 'TKA Soshum']] = df_prov[['Tes Potensi Skolastik', 'TKA Saintek', 'TKA Soshum']].apply(lambda x:round(x,2))
```

```python
df_prov2 = df_prov.join(per_prov.set_index('Provinsi'), on='Provinsi')
df_prov2
```

```python
df_prov2.rename(columns={'Jumlah SMAN': 'Jumlah SMAN Top 1000'}, inplace=True)
df_prov2
```

```python
df2 = pd.read_excel('dataset/Sekolah, Guru, Murid Clean.xlsx')
df2.head()
```

```python
df3 = pd.read_excel('dataset/Jumlah Penduduk Clean.xlsx')
df3.head()
```

```python
df3.drop(['Laki-laki', 'Perempuan'], axis=1, inplace=True)
df3.rename(columns={'Jumlah': 'Jumlah Penduduk (ribu)'}, inplace=True)
```

```python
df3.head()
```

```python
df4 = pd.read_excel('dataset/Luas Daerah clean.xlsx')
df4.head()
```

```python
dataset1 = df_prov2.join(df2.set_index('Provinsi'), on='Provinsi', how='left')
dataset1.head()
```

```python
dataset2 = dataset1.join(df3.set_index('Provinsi'), on='Provinsi', how='left')
dataset2.head()
```

```python
df = dataset2.join(df4.set_index('Provinsi'), on='Provinsi', how='left')
df.head()
```

```python
df['Rate'] = df['Jumlah SMAN Top 1000'] / df['Jumlah Sekolah SMAN'] * 100
df['Rate'] = df['Rate'].apply(lambda x:round(x,2))
df.head()
```

```python
df['Rate Guru'] = df['Jumlah Guru SMAN'] / df['Jumlah Sekolah SMAN']
df['Rate Guru'] = df['Rate Guru'].apply(lambda x:round(x,2))
df.head()
```

```python
df['Rate Murid'] = df['Jumlah Murid SMAN'] / df['Jumlah Sekolah SMAN']
df['Rate Murid'] = df['Rate Murid'].apply(lambda x:round(x,2))
df.head()
```

```python
df['Kepadatan Pendudukan'] = df['Jumlah Penduduk (ribu)'] * 1000 / df['Luas Wilayah (km2)']
df['Kepadatan Pendudukan'] = df['Kepadatan Pendudukan'].apply(lambda x:round(x,2))
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Provinsi'], df['Rate'], color='red', marker='o')
plt.title('Persentase SMAN yang Masuk Top 1000', fontsize=14)
plt.xlabel('Provinsi', fontsize=14)
plt.ylabel('Persentase', fontsize=14)
plt.xticks([])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Provinsi'], df['Rate Guru'], color='red', marker='o')
plt.title('Rerata Guru per Sekolah', fontsize=14)
plt.xlabel('Provinsi', fontsize=14)
plt.ylabel('Jumlah Guru', fontsize=14)
plt.ylim(ymin=0)
plt.xticks([])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Provinsi'], df['Rate Murid'], color='red', marker='o')
plt.title('Rerata Murid per Sekolah', fontsize=14)
plt.xlabel('Provinsi', fontsize=14)
plt.ylabel('Jumlah Murid', fontsize=14)
plt.ylim(ymin=0)
plt.xticks([])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Provinsi'], df['Rate Guru'], marker='o', label="Rerata Guru")
plt.plot(df['Provinsi'], df['Rate'], marker='o', label="Persentase SMAN Top")
plt.title('Rerata Guru dan Persentase SMAN Top 1000', fontsize=12)
plt.xlabel('Provinsi', fontsize=12)
plt.xticks([])
plt.legend()
plt.grid(True)
plt.show()
```
